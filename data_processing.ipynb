```python
# Making the aggregated_transaction dataframe
# importing the necessary libraries
import os
import pandas as pd
import json

# Listing the states directory using os.listdir
path= "D:/PhonePay/data_set/data/aggregated/transaction/country/india/state/"
agg_tran_state= os.listdir(path)
agg_tran_state
agg_tran = {'state':[],'year':[],'quarter':[],'name':[],'count':[],'amount':[]}
for i in agg_tran_state:
    path_i=path+i+"/"
    agg_tran_syear=os.listdir(path_i)
    for j in agg_tran_syear:
        path_j=path_i+j+"/"
        agg_tran_yquater=os.listdir(path_j)
        for k in agg_tran_yquater:
            path_k=path_j+k
            Data=open(path_k,"r")
            D=json.load(Data)
            for l in D['data']['transactionData']:
                Name=l['name']
                Count=l['paymentInstruments'][0]['count']
                Amount=l['paymentInstruments'][0]['amount']
                agg_tran['name'].append(Name)
                agg_tran['count'].append(Count)
                agg_tran['amount'].append(Amount)
                agg_tran['state'].append(i)
                agg_tran['year'].append(j)
                agg_tran['quarter'].append(int(k.strip('.json')))
agg_tran_df=pd.DataFrame(agg_tran)
agg_tran_df.to_csv("D:/PhonePay/dataframes/aggregated_transactions.csv",index=False)
```

```python
# creating the aggregated_insurance dataframe
path_agg_insurance='D:/PhonePay/data_set/data/aggregated/insurance/country/india/state/'
agg_insurance_state=os.listdir(path_agg_insurance)
agg_insurance_state
agg_insurance={
   'State':[],'Year':[],'Quarter':[], 'Count':[],'Amount':[]
}
for ais in agg_insurance_state:
    path_ais=path_agg_insurance+ais+'/'
    agg_insurance_year=os.listdir(path_ais)
    for aiy in agg_insurance_year:
        path_aiy=path_ais+aiy+'/'
        agg_insurance_quater=os.listdir(path_aiy)
        for aiq in agg_insurance_quater:
            path_aiq=path_aiy+aiq
            Data_insurance=open(path_aiq,'r')
            Data_insurance_load=json.load(Data_insurance)
            for dil in Data_insurance_load['data']['transactionData']:
                i_count=dil['paymentInstruments'][0]['count']
                i_amount=dil['paymentInstruments'][0]['amount']
                agg_insurance['Count'].append(i_count)
                agg_insurance['Amount'].append(i_amount)
                agg_insurance['State'].append(ais)
                agg_insurance['Year'].append(aiy)
                agg_insurance['Quarter'].append(int(aiq.strip('.json')))
agg_insurance_df=pd.DataFrame(agg_insurance)
agg_insurance_df.to_csv("D:/PhonePay/dataframes/aggregated_insurance.csv",index=False)
```

```python
# Creating the aggregated user dataframe
path_agg_user='D:/PhonePay/data_set/data/aggregated/user/country/india/state/'
agg_user_state=os.listdir(path_agg_user)
agg_user_state
agg_user={
    'State':[],'Year':[],'Quarter':[],'Brand':[],'Count':[],'Percentage':[]
}
for aus in agg_user_state:
    path_aus=path_agg_user+aus+'/'
    agg_user_year=os.listdir(path_aus)
    for auy in agg_user_year:
        path_auy=path_aus+auy+'/'
        agg_user_quater=os.listdir(path_auy)
        for auq in agg_user_quater:
            path_auq=path_auy+auq
            Data_user=open(path_auq,'r')
            Data_user_load=json.load(Data_user)
            users=Data_user_load['data']['usersByDevice']
            if users is not None:
                for dul in users:
                    agg_user['Brand'].append(dul['brand'])
                    agg_user['Count'].append(dul['count'])
                    agg_user['Percentage'].append(dul['percentage'])
                    agg_user['State'].append(aus)
                    agg_user['Year'].append(auy)
                    agg_user['Quarter'].append(int(auq.strip('.json')))
agg_user_df=pd.DataFrame(agg_user)
agg_user_df.to_csv("D:/PhonePay/dataframes/aggregated_user.csv",index=False)
```

```python
import os
import json
import pandas as pd

# Base path
top_ins_path = "D:/PhonePay/data_set/data/top/insurance/country/india/state/"

# List states
top_ins_states = os.listdir(top_ins_path)

# Data dictionary (same convention)
top_ins = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'District': [],
    'Count': [],
    'Total_Amount': []
}

# Loop through states
for state in top_ins_states:
    state_path = os.path.join(top_ins_path, state)

    # Loop through years
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        # Loop through quarters
        for quarter_file in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter_file)

            # Load JSON
            with open(quarter_path, 'r') as file:
                data = json.load(file)

            # Access districts list safely
            districts_data = data['data'].get('districts', [])

            # Loop through each district entry
            for item in districts_data:
                district_name = item['entityName']
                count = item['metric']['count']
                amount = item['metric']['amount']

                top_ins['State'].append(state)
                top_ins['Year'].append(int(year))
                top_ins['Quarter'].append(int(quarter_file.strip('.json')))
                top_ins['District'].append(district_name)
                top_ins['Count'].append(count)
                top_ins['Total_Amount'].append(amount)

# Create DataFrame
top_ins_df = pd.DataFrame(top_ins)

# Preview
print(top_ins_df.head())

# Save to CSV
top_ins_df.to_csv("D:/PhonePay/dataframes/top_insurance.csv", index=False)
```

```python
import os
import json
import pandas as pd

# Base path
top_user_path = "D:/PhonePay/data_set/data/top/user/country/india/state/"

# List states
top_user_states = os.listdir(top_user_path)

# Data dictionary (same naming convention)
top_user = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'District': [],
    'Registered_Users': []
}

# Loop through states
for state in top_user_states:
    state_path = os.path.join(top_user_path, state)

    # Loop through years
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        # Loop through quarters
        for quarter_file in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter_file)

            # Load JSON
            with open(quarter_path, 'r') as file:
                data = json.load(file)

            # Get districts list
            districts_data = data['data'].get('districts', [])

            # Loop through district records
            for item in districts_data:
                district_name = item['name']
                registered_users = item['registeredUsers']

                top_user['State'].append(state)
                top_user['Year'].append(int(year))
                top_user['Quarter'].append(int(quarter_file.strip('.json')))
                top_user['District'].append(district_name)
                top_user['Registered_Users'].append(registered_users)

# Create DataFrame
top_user_df = pd.DataFrame(top_user)

# Preview
print(top_user_df.head())

# Save to CSV
top_user_df.to_csv(
    "D:/PhonePay/dataframes/top_user.csv",
    index=False
)
```
